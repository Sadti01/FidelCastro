```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import os
os.chdir("/content/drive/My Drive/proyecto2")

from PIL import Image
import urllib.request
import ee # Google Earth Engine

ee.Initialize(project='ee-santiagobautista234')

def sentinel2(start_date, end_date, geometry, imageVisParam):
    IMGSentinel = ee.ImageCollection('COPERNICUS/S2_SR') \
                  .filterDate(start_date, end_date) \
                  .filterBounds(geometry) \
                  .filterMetadata('CLOUDY_PIXEL_PERCENTAGE', 'Less_Than', 30)
    print("size: ", IMGSentinel.size().getInfo())

    collectionList = IMGSentinel.toList(IMGSentinel.size())
    collectionSize = collectionList.size().getInfo()

    imgs = []
    fechas = []
    for i in range(collectionSize):
        SentinelFiltro = ee.Image(collectionList.get(i))
        SentinelClip = SentinelFiltro.clip(geometry)
        imagenRGB = SentinelClip.visualize(**imageVisParam)
        url = imagenRGB.getThumbURL({'dimensions': 3500, 'format': 'png'})
        print(ee.Date(ee.Image(collectionList.get(i)).get('system:time_start')).format('Y-M-d').getInfo())

        imgs.append(imagenRGB)
        fechas.append(ee.Date(ee.Image(collectionList.get(i)).get('system:time_start')).format('Y-M-d').getInfo())
    return (imgs, fechas)

def download_image(list_imagenRGB, rutaImg, extension, view_url=False):
    for i, imagenRGB in enumerate(list_imagenRGB):
        url = imagenRGB.getThumbURL({'dimensions': 3500, 'format': 'png'})
        if view_url:
            print(url)
        # Descargamos la imagen
        try:
            respuesta = urllib.request.urlopen(url)
            image = Image.open(respuesta)
            image.save(rutaImg[i])
            print("Estatus de la descarga: ", True)
        except Exception as e:
            print("Error en la descarga:", e)
            print("Estatus de la descarga: ", False)

# Geometría de ejemplo (Guarne, Colombia)
geometry_guarne = ee.Geometry.MultiPolygon([
    [[[-75.46444905084796, 6.293769979739637],
      [-75.41290772241778, 6.294665769423085],
      [-75.41144860071368, 6.266426323940978],
      [-75.46487820429034, 6.266980889719954],
      [-75.46444905084796, 6.293769979739637]]]
], None, False)

# Parámetros de visualización y fechas
imageVisParam = {"bands": ["B4", "B3", "B2"], "min": 0, "max": 3500}
ext = 'png'
start_date = '2020-01-01'
end_date = '2020-12-28'

# Llamada a las funciones
img_sat, fechas = sentinel2(start_date, end_date, geometry_guarne, imageVisParam)
output_img = ['imagenes/B11_' + fecha.replace('-', '_') + "_" + str(i) + '.' + ext for i, fecha in enumerate(fechas)]
download_image(img_sat, output_img, ext, view_url=True)
```
